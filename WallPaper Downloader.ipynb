```python
import requests
import selenium
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import warnings 
warnings.filterwarnings('ignore')
```

```python
def get_url(keyward):
    keyward = keyward.replace(' ','%20')
    tempurl = 'https://www.pexels.com/search/{}/'
    url  = tempurl.format(keyward)
    return url
```

```python
wall=input('Enter the wallpaper category :')
url = get_url(wall)
# url = get_url('sunflower wallpaper')
print(url)
```

```python
soup = BeautifulSoup(driver.page_source,'html.parser')
all_div = soup.find_all('div','BreakpointGrid_item__erUQQ')
print(len(all_div))
```

```python
firstImgDiv = all_div[0]
testImg = firstImgDiv.find('img')
print(testImg['src'])
```

```python
# Get the SRC of all images 
src=[]
for i in all_div:
    imgTag = i.find('img')
    imgTag = imgTag['src']
    src.append(imgTag)
for i in src:
    print(i)
```
